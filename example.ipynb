```python
### Preamble: imports and stuff like that ###
%matplotlib inline
import TimeSeries
```

```python
n = 100000
armodel = TimeSeries.Generator().generate(n, p=3, alpha = [0.5, 0.3, 0.2], period=n/100)
armodel.plot()
for i in xrange(5):
    armodel.plot(start=i*n/100, end=(i+1)*n/100)
    
```

```python
mamodel = TimeSeries.Generator(type="movingaverage").generate(n, q = 3, theta=[0.5, 0.3, 0.2], mu = 10, period=n/100)
mamodel.plot()
```

```python
for i in xrange(5):
    mamodel.plot(start=i*n/100, end=(i+1)*n/100)
```
